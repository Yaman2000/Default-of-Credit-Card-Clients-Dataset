```python
from scipy import stats

# Calculate the Z-score for each value in the column
credit_card['z_score'] = stats.zscore(credit_card['LIMIT_BAL'])

# Identify outliers (Z-score > 3 or < -3)
credit_card['outlier'] = np.where((credit_card['z_score'] > 3) | (credit_card['z_score'] < -3), True, False)

# Display the data with Z-scores and outliers
print(credit_card)
```
